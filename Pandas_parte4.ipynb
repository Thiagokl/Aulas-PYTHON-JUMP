```python
def update_custo(df):
    categoria = df["Categoria"]
    custo = df["Custo_Unitario"]
    
    ajuste = {
        "Mens Footwear" : 1.11,
        "Bath Clothes" : 1.17,
        "Babywear": 1.22
    }

    aumento = ajuste.get(categoria, 1)

    resultado = round(custo * aumento, 2)

    return resultado
```

# df.head()

```python
def adicionar_des_status(df, status_col, des_status_col):
    status_dict = {
        1.0: 'Ativo', 
        2.0: 'Inativo', 
        3.0: 'Indefinido'
    }
    des_status_list = [status_dict.get(status, 'Desconhecido') for status in df[status_col]]
    
    df[des_status_col] = des_status_list
```

```python
adicionar_des_status(df, 'Status', 'Des_status')
```
