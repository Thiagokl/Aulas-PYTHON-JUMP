- Criando uma função para padronizar o nome das colunas 

````python
def col_capitalize(lista_colunas):
    dicionario_colunas = dict()
    for coluna in lista_colunas:
        dicionario_colunas[coluna] = coluna.capitalize()
    return dicionario_colunas
````
