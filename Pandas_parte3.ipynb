````python
def cria_lista(texto, inicio, fim):
    lista = []
    for i in range(len(texto)):
        result = texto[i][inicio:fim]
        lista.append(result.strip())

    return lista 
    
````

````python
data = cria_lista(texto, 0, 10)
cod_pedido = cria_lista(texto, 10, 16)
cod_cliente = cria_lista(texto, 16, 18)
cod_produto = cria_lista(texto, 18, 22)
volume = cria_lista(texto, 22, 26)
faturamento = cria_lista(texto, 26, 34)
impostos = cria_lista(texto, 34, 41)
````
